```python
df = df.drop_duplicates()
df.reset_index(drop=True,inplace=True)
```

```python
for i in df.columns:
    print(f"Total unique value in {i} column: {df[i].unique()}")
```

```python
#In dataset all features are categorical
```

```python
df['LUNG_CANCER'] = df['LUNG_CANCER'].map({'NO':0, 'YES':1})
```

```python
df['LUNG_CANCER'].unique()
```

```python
df.sample(5)
```

```python
df.dtypes
```

```python
columns = list(df.columns)
columns.remove('AGE')
```

```python
print(columns)
```

```python
for col in columns:
    print(df[col].value_counts(normalize = True)*100)
    print('-'*50)
```

```python
sns.countplot(x=df['GENDER'], hue=df['LUNG_CANCER'], palette='viridis')
plt.title('Count Plot for Gender by Lung Cancer Status')
plt.legend(title='Lung Cancer', labels=['No', 'Yes'], loc='upper left', bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
sns.displot(df['AGE'], kde=True, bins=20, color='blue', height=6, aspect=1.25)
plt.title('Distribution of Age with KDE')
plt.show()
#distribution of age is approximately uniform
```

```python
sns.boxplot(df['AGE'])
plt.show()
#no outlier detected
```

```python
df.columns
```

```python
plt.figure(figsize=(14, 7))
sns.violinplot(x='LUNG_CANCER', y='AGE', data=df, palette='coolwarm')
plt.title('Age Distribution by Lung Cancer Status')
plt.xlabel('Lung Cancer Status')
plt.ylabel('Age')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x=df['SMOKING'], hue=df['LUNG_CANCER'], palette='viridis')
plt.title('Smoking Status vs. Lung Cancer')
plt.xlabel('Smoking Status')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Non-Smoker', 'Smoker'])
plt.legend(title='Lung Cancer', labels=['No', 'Yes'], loc='upper left', bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
for col in columns:
    if col != 'LUNG_CANCER':
        plt.figure(figsize=(12, 6))
        sns.countplot(x=col, hue='LUNG_CANCER', data=df, palette='deep')
        plt.title(f'{col} vs. Lung Cancer')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.legend(title='Lung Cancer', labels=['No', 'Yes'])
        plt.show()
```

```python
plt.figure(figsize=(14, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
